# DATA CLEANING/ FEATURE ENGINEERING

```python
df_alexa=df_alexa.drop(['date','rating'],axis=1)
```

```python
df_alexa
```

```python
variation_dummies=pd.get_dummies(df_alexa['variation'],drop_first=True).astype(int)
variation_dummies
```

```python
df_alexa.drop(['variation'],axis=1,inplace= True)
df_alexa
```

```python
df_alexa=pd.concat([df_alexa,variation_dummies],axis=1)
df_alexa
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
nan_values=df_alexa[df_alexa['verified_reviews'].isna()]
```

```python
nan_values
```

```python
df_alexa=df_alexa.dropna(subset=['verified_reviews'])
```

```python
vectorizer= CountVectorizer()
alexa_countvectorizer= vectorizer.fit_transform(df_alexa['verified_reviews'])
```

```python
alexa_countvectorizer.shape
```

```python
np.set_printoptions(threshold=np.inf)
print(vectorizer.get_feature_names_out())
```

```python
print(alexa_countvectorizer.toarray()[:10])
```

```python
df_alexa.drop(['verified_reviews'],axis=1,inplace=True)
```

```python
df_alexa
```

```python
encoded_reviews=pd.DataFrame(alexa_countvectorizer.toarray())
```

```python
df_alexa=pd.concat([df_alexa,encoded_reviews],axis=1)
```

```python
df_alexa
```

```python
df_alexa.dropna(inplace=True)
```

```python
df_alexa
```

```python
X=df_alexa.drop(['feedback'],axis=1)
```

```python
X.shape
```

```python
y=df_alexa['feedback']
```

```python
y
```

# MODEL TRAINING

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
randomforest_classifier= RandomForestClassifier(n_estimators=50,criterion='entropy')
```

```python
X_train.columns = X_train.columns.astype(str)
```

```python
randomforest_classifier.fit(X_train,y_train)
```

```python
y_predict_train=randomforest_classifier.predict(X_train)
```

```python
y_predict_train
```

```python
y_train
```

```python
cm=confusion_matrix(y_train,y_predict_train)
```

```python
sns.heatmap(cm, annot=True)
plt.show()
```

```python
print(classification_report(y_train,y_predict_train))
```

```python
X_test.columns = X_test.columns.astype(str)
y_predict=randomforest_classifier.predict(X_test)
```

```python
cm=confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cm, annot=True)
plt.show()
```

```python
print(classification_report(y_test,y_predict))
```

# IMPROVE THE MODEL

```python
df_alexa=pd.read_csv('amazon_alexa.tsv',sep='\t')
```

```python
df_alexa=pd.concat([df_alexa,pd.DataFrame(alexa_countvectorizer.toarray())],axis=1)
```

```python
df_alexa
```

```python
df_alexa['verified_reviews'] = df_alexa['verified_reviews'].astype(str)

df_alexa['length']=df_alexa['verified_reviews'].apply(len)
```

```python
df_alexa
```

```python
df_alexa.dropna(inplace=True)
```

```python
X=df_alexa.drop(['rating','date','variation','verified_reviews','feedback'],axis=1)
```

```python
X
```

```python
y=df_alexa['feedback']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
randomforest_classifier= RandomForestClassifier(n_estimators=50,criterion='entropy')
```

```python
X_train.columns = X_train.columns.astype(str)
```

```python
randomforest_classifier.fit(X_train,y_train)
```

```python
y_predict_train=randomforest_classifier.predict(X_train)
```

```python
cm=confusion_matrix(y_train,y_predict_train)
```

```python
sns.heatmap(cm, annot=True)
plt.show()
```

```python
print(classification_report(y_train,y_predict_train))
```

```python
X_test.columns = X_test.columns.astype(str)
y_predict=randomforest_classifier.predict(X_test)
```

```python
y_predict
```

```python
cm=confusion_matrix(y_test,y_predict)
sns.heatmap(cm, annot=True)
plt.show()
```

```python
print(classification_report(y_test,y_predict))
```
